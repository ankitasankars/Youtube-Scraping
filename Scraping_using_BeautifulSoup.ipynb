## Using BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests

HEADER = {
    "User-Agent": 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}
link = "https://youtu.be/C-gEQdGVXbk"
```

#### Turning the HTML content of the given link into a structured tree.

```python
response = requests.get(link, headers=HEADER)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

#### Title of the Video

```python
titleSoupMeta = soup.find("meta", property="og:title")
videoTitle = titleSoupMeta["content"] if titleSoupMeta else "NotFound"

print(videoTitle) 
```

#### Video Thumbnail

```python
thumbSoupMeta = soup.find("meta", property="og:image")
videoImage = thumbSoupMeta["content"] if thumbSoupMeta else "NotFound"

print(videoImage)
```

#### Number of Views

```python
videoViews = soup.find("meta", itemprop="interactionCount")['content']
```

#### Description of the Video

```python
videoDescription = soup.find("meta", itemprop="description")['content']
```

#### Video's Publishing Date

```python
videoPublished = soup.find("meta", itemprop="datePublished")['content']
```

#### Tags used for the video

```python
videoTags = ', '.join([ meta.attrs.get("content") for meta in soup.find_all("meta", {"property": "og:video:tag"}) ])
```

#### Likes and Dislikes

```python
likeButon = soup.find(attrs={"class": "like-button-renderer-like-button"})
dislikeButon = soup.find(attrs={"class": "like-button-renderer-dislike-button"})
```

#### Final Data

```python
print("Title =", videoTitle)
print("Thumbnail =", videoImage)
print("Total Views =", videoViews)
print("Publish Date =", videoPublished)
print("Tags =", videoTags)
from pytube import YouTube
try:
    yt_obj = YouTube('https://youtu.be/C-gEQdGVXbk')
    print(f'Video Length = {yt_obj.length} seconds')
    print(f'Description = {yt_obj.description}')
    print(f'ideo Rating = {yt_obj.rating}')
    print(f'Video Author = {yt_obj.author}')
except Exception as e:
    print(e)
```
